Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset


```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

Univariate analysis


```python
sns.histplot(tips['total_bill'],kde=True)
plt.title("Distribution of total bill")
plt.show()
```

The distribution is slightly rightly skewed.Most bills are between $10-$20.

```python
sns.countplot(x='day',data=tips)
plt.title('Count of tips per day')
plt.show()
```

Saturdays and sundays have the highest number of customers indicating weekends are bussier.
Friday has least traffic .

Bivariate analysis

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
```

From the scatter plot it shows a rise in the tips with increse of bill size.Hence, grater the bill larger the tip.

```python
sns.boxplot(x='day', y='tip', data=tips)
```

Sundays show a slightly higher spread and a few higher ooutlier tips

```python
sns.pairplot(tips,hue='sex')
```

total_bill vs tip
Clear positive linear trend: higher bill  higher tip.

Both genders follow a similar trend, but males seem to appear more frequently with slightly higher bills.

 total_bill vs size
Slight upward trend: larger parties often have larger bills.

However, it's not strictly linear (e.g., group size 2 with a wide bill range).

tip vs size
Weak positive trend: tips slightly increase with group size.

But many tips for small groups overlap with those from large groups.

```python
# Only use numeric columns for correlation
numeric_data = tips.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

Total_bill and size have a 0.6 corelation,total_bill and tip=0.6 .
Larger groups speand moderately but dont tip much.Tips increase with increase in total bill.
